```python


# called 'df'
file_path = 'player_information.csv'
df = pd.read_csv(file_path)

# birth city column is named 'player_birthcity'
birth_city_column = 'player_birthCity'

# API base URL
api_base_url = 'https://api.api-ninjas.com/v1/city?'

#  API key
api_key = 'nVSl3APKDcl75e5gd1nmBg==GekBJP6bGG8nSITj'

# Lists to store latitude and longitude values
latitudes = []
longitudes = []

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    birth_city = row[birth_city_column]
    
    # Construct the API URL
    api_url = api_base_url + 'name={}'.format(birth_city)
    
    # Make the API request
    response = requests.get(api_url, headers={'X-Api-Key': api_key})
    
    if response.status_code == requests.codes.ok:
        data_list = response.json()
        if data_list:
            data = data_list[0]  # Assuming the API returns a list of results and we take the first one
            latitude = data.get('latitude')
            longitude = data.get('longitude')
            latitudes.append(latitude)
            longitudes.append(longitude)
        else:
            latitudes.append(None)
            longitudes.append(None)
    else:
        latitudes.append(None)
        longitudes.append(None)

# Add latitude and longitude lists as new columns to the DataFrame
df['Latitude'] = latitudes
df['Longitude'] = longitudes

```

```python
# Reorder columns and rename them
new_column_order = ['player_id','player_name','player_firstName','player_lastName','player_birthDate','player_birthCity','Latitude','Longitude','player_birthCountry','player_birthStateProvince','player_nationality','player_height','player_weight','player_shootsCatches','player_primaryPositionCode']
df = df[new_column_order]

# Rename columns
new_column_names = {
    'Latitude': 'player_birthCity_Lat',
    'Longitude': 'player_birthCity_Long',
   
}
player_info_LatLong_df = df.rename(columns=new_column_names)
player_info_LatLong_df.info()
```

```python
# Save the modified DataFrame to a new CSV file
new_file_path = 'player_information_with_lat_long.csv'
player_info_LatLong_df.to_csv(new_file_path, index=False,encoding = 'utf-8-sig')

print(f"Modified DataFrame saved to {new_file_path}")
```
